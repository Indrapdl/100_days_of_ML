```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
inp = pd.read_csv('/content/placement.csv')
```

```python
inp.head()
```

```python
#eda to understand data
#plotting
plt.scatter(inp['cgpa'],inp['iq'],c=inp['placement'])
```

```python
#extracting input and output column separately
x = inp.iloc[:,0:2]
y = inp.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
x.shape
```

```python
y.shape
```

```python
#splitting into training and testing
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size = 0.1)
```

```python
X_train
```

```python
X_test
```

```python
Y_train
```

```python
Y_test
```

```python
#scaling data
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_train
```

```python
X_test = scaler.fit_transform(X_test)
```

```python
X_test
```

```python
#Now training the model
#logistic regression(In this case)
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression()
```

```python
#fitting logistic regression model
clf.fit(X_train,Y_train)
```

```python
#evaluating model
Y_pred = clf.predict(X_test)
```

```python
Y_test
```

```python
#calculating accuraccy score
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(Y_test,Y_pred)
accuracy
```

```python
#plotting decision boundary
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(X_train, Y_train.values,clf=clf, legend=2)
```

```python
#exporting the  model(for other use) export model as file
import pickle
pickle.dump(clf,open('model.pkl','wb'))
```
